```python
#We are excluding id & host_id because they are unique ids, lat long since they are coordinates.
column_after_exclude= set(ds.columns) - {'latitude','longitude','id','host_id'}
ds[column_after_exclude].describe()
```

```python
dist_column_list = ds[column_after_exclude].describe().columns.tolist()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(27, 16))
axes = axes.flatten()
for col, ax in zip(dist_column_list, axes):
    sns.histplot(x=col, data=ds, ax=ax, kde=True, element='poly')
    ax.set_title(f'Column {col} skewness : {ds[col].skew()}')

plt.tight_layout(h_pad=0.6, w_pad=0.9)
```

It is clear from the distribution of the columns of filtered numerical data that all of them, including price, have a positively skewed distribution. However, because availability is distributed uniformaly throughout the days of the year, we have all kinds of listings readily available all year round.

```python
plt.figure(figsize=(11, 7))
heatmap = sns.heatmap(ds[dist_column_list].corr(), linewidths=0, vmin=-1, annot=True, cmap="YlGnBu")
plt.show()
```

For obvious reasons, the number of reviews and reviews per month have a high correlation. However, there is relatively low correlation between the price column and other attributes. let's Check out the differences and connections.

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(26, 14))
ax = axes.flatten()

sns.set_theme(style="white")
sns.scatterplot(data=ds, x='longitude', y='latitude', hue='neighbourhood_group', ax=ax[0]);
ax[0].set_title('Location of neighbourhood groups')
sns.scatterplot(data=ds[ds['price'] < 300], x='longitude', y='latitude', hue='price', size="price", sizes=(20, 60), palette='GnBu_d', ax=ax[1])
ax[1].set_title('Variation of Price based on Location ($0 - 300)')
sns.scatterplot(data=ds, x='longitude', y='latitude', hue='number_of_reviews', size="number_of_reviews", sizes=(20, 150), palette='GnBu_d', hue_norm=(0, 5), ax=ax[2])
ax[2].set_title('Variation of number of reviews given based on location')
sns.scatterplot(data=ds, x='longitude', y='latitude', hue='availability_365', style="room_type", palette='GnBu_d', ax=ax[3])
ax[3].set_title('Availability in terms of Room Type')
plt.show()
```

Let's discuss about the above plot

* The first plot allows us to check which New York City neighbourhood currently has our dataset.
* Given that our 75th percentile data falls within a range of USD 175 in the second plot, we only evaluated listings with a price range maxing out at USD 300. We can examine the distribution of price variance across the city. The pricey parts of New York are located south of Manhattan and north of Brooklyn. (Timeout.com, source.
* In the third plot, we can observe a pattern of an increase in the number of reviews on city outskirts.
* We attempted to depict the availability in terms of room type in the previous plot.Though availability based on room type is spreaded well, still we can follow a pattern where the heart of new york stays the busiest or booked for most of the time.

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(26, 14))
ax = axes.flatten()

# group_price_ds = ds.groupby('neighbourhood_group')['price'].agg([np.max, np.mean, np.min]).reset_index()
mean_price_ds = ds.groupby(['neighbourhood_group', 'room_type'], as_index=False)[['price']].mean()
sns.barplot(data=mean_price_ds, x='neighbourhood_group', y ='price', hue='room_type', ax=ax[0])

sns.histplot(data=ds[ds['price'] < 500], x="price", hue="neighbourhood_group", stat='frequency', element='poly', ax=ax[1])
sns.histplot(data=ds, x='neighbourhood_group', y='price', hue='room_type', element='poly', ax=ax[2])
ax[1].set_title('Distribution of Listing and Price throughout city.')
sns.despine(fig)
```

In terms of listing areas and cost, it is clear from the accompanying graph that Brooklyn and Manhattan are located in the most populated and vibrant location. The most expensive average hotel price are in Manhattan and Brooklyn, however Staten Island is not far behind. In every community, shared rooms are generally inexpensive and few in number; nonetheless, whilst Manhattan has most number of Entire home/apt category, but Brooklyn has most number of Private room category.

Analyse this in light of room category.

```python
fig = plt.figure(figsize=(26, 8))
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)
sns.lineplot(data=ds, x='neighbourhood', y='price', hue='room_type')
plt.title('Distribution of Prices in neighbourhoods')
sns.despine(fig)
```

Obviously entire room type practically in all communities,the price range for entire homes and apartment has remain higher.

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(24, 9))
ax = axes.flatten()

mean_price_ds = ds.groupby('room_type', as_index=False)[['price']].mean()
sns.barplot(data=mean_price_ds, x='room_type', y='price', palette='GnBu_d', ax=ax[0])
ax[0].set_title("Avg Price vs Rooms type");

labels = ds['room_type'].value_counts().index
sizes = ds['room_type'].value_counts().values
ax[1].pie(sizes, labels=labels, autopct='%1.1f%%', colors = ['#009999','#007399','#20B2AA'])
ax[1].set_title('Proportion of Room Types')

sns.countplot(data=ds, x='room_type', ax=ax[2])
ax[2].set_title('Room Type Counts')

sns.despine(fig)
plt.tight_layout(h_pad=0.5, w_pad=0.8)
```

In New York City, the percentage of entire homes/apartments is higher than 55%, and their average price is likewise the highest.Shared room are the cheapest, but only has 2.4% proportion. No wonder New York life is of high standard.

Now we will look into availability varies with all neighbourhoods, room standards.

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(26, 9))
ax = axes.flatten()

sns.lineplot(data=ds, x='neighbourhood_group', y='availability_365', hue='room_type', ax=ax[0])
ax[0].set_title('Room Availability throughout Neighbourhood/Room Type')
sns.scatterplot(data=ds[ds['price'] < 500], x="availability_365", y='price', hue='room_type', alpha=.9, palette="muted", ax=ax[1])
ax[1].set_title('Price vs Availability (Range $ 0 - 500)')

sns.countplot(data=ds[ds['availability_365']  == 365], x='neighbourhood_group', hue='room_type', palette='GnBu_d', ax=ax[2])
ax[2].set_title('Property Available 365 days')
sns.despine(fig)
```

* Staten Island has the busiest Shared Room and the most available Private Room, according to the two graphs above. The availability in Brooklyn and Manhattan is almost similar.
* On the other hand, the Price vs. Availability graph shows a very uniform distribution, meaning that a specific room type in a specific price range can be booked all year round.
* Also, there are 1295 homes in total that are accessible 365 days a year. With 572 properties, Manhattan comes in first, followed by Brooklyn with 453 homes.

```python
ds['last_review'] = pd.to_datetime(ds['last_review'], errors='coerce')

ds['Days'], ds['Month'], ds['Years'] = (ds['last_review'].dt.day, ds['last_review'].dt.month, ds['last_review'].dt.year)
ds['last_review'] = pd.to_datetime(ds['last_review']).dt.date
```

```python
#As we are eliminating these data because after the conversion to date, the null column date became the system minimum date, i.e 1970-01-01.
filtered_ds = ds[ds['Years']!=1970]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(22, 6))
ax = axes.flatten()

sns.histplot(data=filtered_ds, x='Days', hue='room_type', multiple="stack", ax=ax[0])
ax[0].set_title('No of Guest in days of the months')
sns.histplot(data=filtered_ds, x='Month', hue='room_type', multiple="stack", ax=ax[1])
ax[1].set_title('No of Guest in months of the years')
sns.despine(fig, left=True)
```

Since many guests opt not to provide a rating, we have created the day and month based on the last_review date, however this is only a guess. We can see a pattern where the first and last days of the month have the most visitors. The number of visitors increases in June, which is also in the middle of the year.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(24, 9))
ax = axes.flatten()

sns.lineplot(data=ds, x='neighbourhood_group', y='reviews_per_month', hue='room_type', style="room_type",
        markers=True, dashes=False, ax=ax[0])
ax[0].set_title('Review per Month vs Neighbourhoods')
sns.scatterplot(data=ds, x='number_of_reviews', y='price', hue='room_type', palette="ch:r=-.2,d=.3_r",
        hue_order=['Private room', 'Shared room', 'Entire home/apt'], ax=ax[1])
ax[1].set_title('Price vs Number of Reviews')
sns.despine(fig, left=True)
```

When compared to Manhattan and Brooklyn, the monthly review rate is fairly low. Also we can notice a bad correlation between pricing & reviews. Whereas more affordable properties have a lot of reviews, while more expensive properties have a lot fewer. We know that more people visit cheaper rooms than more expensive ones and that the quantity of reviews is directly inversely related to the number of visitors.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(24, 9))
ax = axes.flatten()

mean_price_ds = ds.groupby(['neighbourhood_group', 'room_type'], as_index=False)[['minimum_nights']].mean()
sns.barplot(data=mean_price_ds, x='neighbourhood_group', y='minimum_nights', hue='room_type', palette='GnBu_d', ax=ax[0])
ax[0].set_title("Avg Minimum Nights vs Neighbourhood / Room Type");

sns.scatterplot(data=ds[ds['price'] < 1000], x='minimum_nights', y='price', hue='room_type', palette="ch:r=-.2,d=.3_r", ax=ax[1])
ax[1].set_title("Price vs Minimum Nights (Range $ 0 - 1000)")

sns.despine(fig)
```

As we can see, properties tend to allow fewer minimum nights the more expensive they are. The minimum nights distribution is ranged between 0 and 400 days for the price range of 0 to 200. However, the average minimum nights reduced to 7 nights within the $600–$1000 price range. But even so, we can claim that the majority of hosts require a stay of fewer than ten nights.

As a trivia, Manhattan has highest avg minimum night restriction.
